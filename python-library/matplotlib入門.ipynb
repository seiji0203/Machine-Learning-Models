# Matplotlib入門


**Matplotlibの基礎を学ぼう**<br>
機械学習において、アルゴリズムから最善の結果を得ようとした場合、多くの場合でデータセットの深い理解が必要となります。データを理解する方法として、最も手っ取り早いのが「データの可視化」です。そのデータの可視化を効率的かつ簡単に行えるのが、Matplotlibとなります。多くの場合で、機械学習で扱うデータの量は膨大で、一つ一つのデータを確認するのはナンセンスです。大量のデータを紐解く便利なツールとして、Matplotlibの基本的な使い方を覚えましょう。本コース「Matplotlib 入門」では、Numpyも併せて使います。


**Matplotlibとは？**<br>
読み方は「マットプロットリブ」です。Matplotlibは、PythonとNumpy（科学計算用ライブラリー）のためのグラフ描写ライブラリーです。基本的な折れ線グラフからヒストグラムなど、様々なグラフを簡単に描写することが可能です。機械学習においてMatplotlibは便利なツールの一つです。機械学習のほぼすべての工程において、Matplotlibを使うことで効率よくデータを理解することが可能です。データセットをどれくらい深く理解して、巧く活用するかが機械学習の核となります、その工程を大きく手助けしてくれるのがMatplotlibなのです。Matplotlibのすべての使い方を覚えるのは非効率ですが、基本的な使い方を覚えておけば、その都度、必要に応じて公式ドキュメントなどで調べながら、データの解析を行うことが可能です。

# グラフの凡例とタイトル
グラフの凡例（Legend）、タイトル、目盛線 （グリッド）の操作を行ってみましょう。また、グラフ描写時に関数を使ってプロッティングを行ってみましょう。

* plt.legend(loc='upper left') 凡例を左上部へ表示
* plt.title() グラフのタイトル指定
* plt.grid(True) グラフのグリッド表示

```python
# データ生成
x = np.linspace(0, 25, 100)
x
```

```python
# plt.legend(loc='upper left') 凡例を左上部へ表示
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, "-b", label="sine")
plt.plot(x, y2, "-r", label="cosine")
plt.legend(loc='upper left')
plt.ylim(-1.5, 2.0)
plt.show()
```

```python
# plt.title() グラフのタイトル指定
# plt.grid(True) グラフのグリッド表示
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, "-b", label="sine")
plt.plot(x, y2, "-r", label="cosine")
plt.legend(loc='upper left')
plt.title("Sine & Cosine Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.ylim(-1.5, 2.0)
plt.show()
```

```python
def sq(x):
  return x**2
def cube(x):
  return x**3
def power4(x):
  return x**4

plt.plot(x, sq(x), 'r')
plt.plot(x, cube(x), 'b')
plt.plot(x, power4(x), 'g')

plt.show()
```

# サブプロット

https://kino-code.com/python-matplotlib-subplot/

```python
x1 = [1, 2, 3, 4, 5]
y1 = [1, 5, 2, 4, 3]
x2 = [1, 2, 3, 4, 5]
y2 = [5, 1, 4, 2, 3]
```

# ヒストグラム（度数分布図）をMatplotlibで描写

* 標準正規分布 = Standard Normal Distribution
* 標準正規分布は平均が0、分散が1の正規分布
* 連続一様分布 = Continuous Uniform Distribution
* np.random.randn() 標準正規分布に属する乱数生成
* np.random.random() 連続一様分布に属する乱数生成
* plt.hist() ヒストグラムの作成

```python
 #標準正規分布 = Standard Normal Distribution
 g = np.random.randn(1000)
```

```python
 g.mean()
```

```python
g.std()
```

```python
plt.hist(g)
plt.show()
```

```python
plt.hist(g, bins=100)
plt.show()
```

```python
# rangeを指定
plt.hist(g, bins=100, range=(0, 5))
plt.show()
```

```python
# 累積値
plt.hist(g, bins=100, cumulative=True)
plt.show()
```

```python
# histtypeを変更
plt.hist(g, bins=100, rwidth=0.5, color='red', histtype="step")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 20)

y1 = np.sin(x)
y2 = 12*x + 8
y3 = x**2
y4 = np.exp(x)

figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

axes[0,0].plot(x, y1,"r")
axes[0,1].plot(x, y2,"b")
axes[1,0].plot(x, y3,"g")
axes[1,1].plot(x, y4,"-")
figure.show()
```
