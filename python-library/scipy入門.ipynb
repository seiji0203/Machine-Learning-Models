```python
# モジュールのインポート
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
plt.style.use("ggplot")
np.random.seed(123)
```

```python
# 若干ノイズを混ぜたデータを生成
x = np.arange(20)
y = 2.0 * x
y_noise = np.random.randn(20)
y = y + y_noise
plt.scatter(x, y)
```

```python
param = np.polyfit(x, y, 1)
param
```

```python
# 多項式近似
plt.plot(x, param[0]*x + param[1])
plt.scatter(x, y)
```

```python
# 指数関数的に発散するデータ生成
x2 = np.arange(15)
y2 = np.exp(0.4 * x2)
y2 = y2 + 10.0 * np.random.randn(15)
plt.scatter(x2, y2)
```

```python
# 非線形関数を定義
def nonlinear_fit(x, a, b, c):
    return a * np.exp(b * x) + c
```

```python
popt, pcov = curve_fit(nonlinear_fit, x2, y2)
plt.scatter(x2, y2)
plt.plot(x2, nonlinear_fit(x2, popt[0], popt[1], popt[2]))
```

# Scipyによる数値積分

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy import integrate
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
x = np.arange(-100, 100, 0.01)
y = np.exp(-x**2)
plt.plot(x, y)
plt.xlim([-5,5])
```

```python
integrate.simps(y, dx=0.01)
```

$\int_{-\infty}^{\infty}\exp(-ax^2)dx=\sqrt{\frac{\pi}{a}}$

```python
np.sqrt(np.pi)
```

```python
# cumtrapzでは積分の過程も確認できます
integrate.cumtrapz(y, dx=0.01)
```
